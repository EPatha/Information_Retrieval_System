```python
# Step 1: Import necessary libraries
import pandas as pd
from google.colab import files

# Step 2: Upload the file
uploaded = files.upload()

# Step 3: Load the dataset
# Assuming 'output_catur_white.xlsx' is the file you uploaded
file_path = list(uploaded.keys())[0]  # Gets the name of the uploaded file
data = pd.read_excel(file_path)

# Step 4: Display the first few rows to understand the structure
data.head()
```

```python
# Step 4: Filter games where the first move by White is 'e4', 'c4', or 'b3'
filtered_data = data[data['Moves'].str.startswith(('e4', 'c4', 'b3'))]

# Map results to numerical values for win rate calculation
filtered_data['Result'] = filtered_data['Result'].map({'1-0': 1, '0-1': 0, '1/2-1/2': 0.5})
filtered_data.head()
```

```python
# Step 5: Calculate total games and win rate for each opening move
win_rate = filtered_data.groupby(filtered_data['Moves'].str[:2]).agg(
    total_games=('Result', 'count'),
    win_rate=('Result', 'mean')
).reset_index()

# Rename columns for clarity
win_rate.columns = ['Opening Move', 'Total Games', 'Win Rate']
win_rate
```
